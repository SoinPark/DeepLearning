# RESNet 응용하기

```python
from tensorflow.keras.applications.resnet50 import ResNet50, decode_predictions
resnet = ResNet50()
resnet.summary()
```

```python
import numpy as np
from PIL import Image
```

```python
img = np.array(Image.open('images/cat.jpg').resize((224,224)))

yhat = resnet.predict(img.reshape(-1, 224, 224, 3))
label = decode_predictions(yhat)
label = label[0][0]
print(f'{label[1]} ({label[2]*100:.2f}%)')
```
